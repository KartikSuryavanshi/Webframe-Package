## 2. Import the Package

Import the main functions from the package:

```python
from webframe import render_site, render_youtube_video

print("✅ Package imported successfully!")
```

## 3. Test Website Embedding

Let's embed some websites directly in the notebook:

```python
# Example 1: Render Python Documentation
render_site("https://docs.python.org/3/")
```

```python
# Example 2: Render GitHub with custom dimensions
render_site("https://github.com", width=1000, height=700)
```

## 4. Test YouTube Video Embedding

Now let's embed some YouTube videos:

```python
# Example 3: Embed a Python tutorial video
render_youtube_video("https://www.youtube.com/watch?v=rfscVS0vtbw")
```

```python
# Example 4: Custom size video
render_youtube_video("https://youtu.be/rfscVS0vtbw", width=1000, height=600)
```

## 5. Test Error Handling

Let's test how the package handles invalid inputs:

```python
# Test 1: Invalid URL
try:
    render_site("not-a-valid-url")
except Exception as e:
    print(f"✅ Caught expected error: {type(e).__name__}")
    print(f"   Message: {e}")
```

```python
# Test 2: Invalid YouTube URL
try:
    render_youtube_video("https://not-youtube.com/watch?v=123")
except Exception as e:
    print(f"✅ Caught expected error: {type(e).__name__}")
    print(f"   Message: {e}")
```
